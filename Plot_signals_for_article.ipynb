```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
archivo_path_1 = os.path.join(os.getcwd(), 'data2', f'or_s_1.npy')
archivo_path_2 = os.path.join(os.getcwd(), 'data2', f'int_s_1.npy')
archivo_path_3 = os.path.join(os.getcwd(), 'data2', f'sg_s_1.npy')
archivo_path_4 = os.path.join(os.getcwd(), 'data2', f'sw_s_1.npy')
archivo_path_5 = os.path.join(os.getcwd(), 'data2', f'flk_s_1.npy')
archivo_path_6 = os.path.join(os.getcwd(), 'data2', f'transient_s_1.npy')
archivo_path_7 = os.path.join(os.getcwd(), 'data2', f'hrc_s_1.npy')
archivo_path_8 = os.path.join(os.getcwd(), 'data2', f'hrc_sg_s_1.npy')
archivo_path_9 = os.path.join(os.getcwd(), 'data2', f'hrc_sw_s_1.npy')
archivo_path_10 = os.path.join(os.getcwd(), 'data2', f'hrc_int_s_1.npy')

# Cargar el archivo completo
signal_data_1 = np.load(archivo_path_1)
signal_data_2 = np.load(archivo_path_2)
signal_data_3 = np.load(archivo_path_3)
signal_data_4 = np.load(archivo_path_4)
signal_data_5 = np.load(archivo_path_5)
signal_data_6 = np.load(archivo_path_6)
signal_data_7 = np.load(archivo_path_7)
signal_data_8 = np.load(archivo_path_8)
signal_data_9 = np.load(archivo_path_9)
signal_data_10 = np.load(archivo_path_10)

fig, axs = plt.subplots(5, 2,figsize=(15,15))
axs[0, 0].plot(signal_data_1,linewidth=0.7)
axs[0, 0].set_title('Original Signal',weight='bold')
axs[0, 1].plot(signal_data_2,linewidth=0.7)
axs[0, 1].set_title('Interruption Signal',weight='bold')
axs[1, 0].plot(signal_data_3,linewidth=0.7)
axs[1, 0].set_title('Sag Signal',weight='bold')
axs[1, 1].plot(signal_data_4,linewidth=0.7)
axs[1, 1].set_title('Swell Signal',weight='bold')
axs[2, 0].plot(signal_data_5,linewidth=0.7)
axs[2, 0].set_title('Flicker Signal',weight='bold')
axs[2, 1].plot(signal_data_6,linewidth=0.7)
axs[2, 1].set_title('Transient Signal',weight='bold')
axs[3, 0].plot(signal_data_7,linewidth=0.7)
axs[3, 0].set_title('Harmonic Signal',weight='bold')
axs[3, 1].plot(signal_data_8,linewidth=0.7)
axs[3, 1].set_title('Harmonic+Sag Signal',weight='bold')
axs[4, 0].plot(signal_data_9,linewidth=0.7)
axs[4, 0].set_title('Harmonic+Swell Signal',weight='bold')
axs[4, 1].plot(signal_data_10,linewidth=0.7)
axs[4, 1].set_title('Harmonic+Interruption Signal',weight='bold')



for ax in axs.flat:
    ax.set(xlabel='Muestras', ylabel='Amplitud')
    plt.subplots_adjust(left=None, bottom=3, right=None, top=4, wspace=0.2, hspace=0.4)


plt.show()
```
